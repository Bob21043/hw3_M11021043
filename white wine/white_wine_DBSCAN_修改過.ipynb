```python
import pandas as pd
import time
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
```

```python
df=pd.read_csv("/content/drive/MyDrive/資料探勘/作業2/hw3/Wine Quality Data Set.csv")
df
```

```python
scaler=MinMaxScaler()
df_1=scaler.fit_transform(df)
df_2=pd.DataFrame(df_1,columns=df.columns)
df_2
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn import datasets
```

```python
Start = time.time()
DBSCAN = DBSCAN(eps=0.35,min_samples=10).fit(df_2)
DBSCAN_label=DBSCAN.fit_predict(df_2)
print("results:")
print(DBSCAN_label)
End = time.time()
print("運算時間: ",End-Start)
```

```python
from sklearn.metrics import accuracy_score
```

```python
len(DBSCAN_label)
```

```python
DBSCAN_add = df_2
DBSCAN_add["Cluster"] = DBSCAN_label
DBSCAN_add
```

```python
DBSCAN_results = DBSCAN_add.groupby("Cluster").sum()
DBSCAN_results
```

```python
purityList=[]
total=0
Total_purity=0
for i in range(1):
    purity = DBSCAN_results.loc[i].max()/DBSCAN_results.loc[i].sum()
    purityList.append(purity)
    total = total + DBSCAN_results.loc[i].sum()
    Total_purity = Total_purity+purityList[i]*(DBSCAN_results.loc[i].sum()/total)

print("純度:\n",purityList)
print("總和:\n",total)
print("總純度:\n",Total_purity)
```
